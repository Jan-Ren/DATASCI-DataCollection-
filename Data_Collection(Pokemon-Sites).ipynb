```python
###Import "requests" library 
```

```python
import requests
import json
URL="https://bulbapedia.bulbagarden.net/wiki/List_of_Pok%C3%A9mon_by_National_Pok%C3%A9dex_number"
```

```python
###Get page content
```

```python
page = requests.get(URL)
```

```python
###Parse HTML file
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(page.content, 'html.parser')
```

```python
###Get the main content for tables
poke_content=soup.find(id='mw-content-text')

###Get Generation Number
poke_generations=poke_content.find_all('h3')

###Get tables from the content
poke_tables=poke_content.find_all('table')

###Check poke_tables for Generation I to Generation VIII and check how to get links
link = poke_tables[8].contents[3].find_all('a')
link[1].get('href')
```

```python
###Loop to all generation and check table. Generate JSON file
gen_start = 1
gen_end = 9

list_of_all_gens = []
gen_json = []

info_start = 3
for i in range(gen_start, gen_end):
    curr_table = poke_tables[i]
    gen = poke_generations[i-1].text.strip()
    if i == 1:
        pokedex = "kdex";
    elif i == 2:
        pokedex = "jdex";
    elif i == 3:
        pokedex = "hdex";
    elif i == 4:
        pokedex = "sdex";
    elif i == 5:
        pokedex = "udex";
    elif i == 6:
        pokedex = "kdex";
    elif i == 7:
        pokedex = "adex";
    elif i == 8:
        pokedex = "gdex";
        
    for j in range(info_start, len(curr_table.contents), 2):
        poke_info=curr_table.contents[j]
        kdex=poke_info.contents[1].text.strip()
        ndex=poke_info.contents[3].text.strip()
        name=poke_info.contents[7].text.strip()
        type1=poke_info.contents[9].text.strip()
        link = poke_info.find_all('a')
        URL = link[1].get('href')
        if len(poke_info.contents) > 10:
            type2=poke_info.contents[11].text.strip()
            gen_json.append({
                pokedex: kdex,
                "ndex": ndex,
                "name": name,
                "type1": type1,
                "type2": type2,
                "gen": gen,
                "URL": URL
            })
        else:
            gen_json.append({
                pokedex: kdex,
                "ndex": ndex,
                "name": name,
                "type1": type1,
                "gen": gen,
                "URL": URL
            })  
    list_of_all_gens.append(gen_json)
    gen_json = []
```

```python

###Get each generation json format, n = 0 (1st gen) to n = 7 (8th gen)
### Generation I
print(list_of_all_gens[0])
with open('generation1.json', 'w', encoding='utf-8') as f:
    json.dump(list_of_all_gens[0], f, ensure_ascii=False, indent=4)
```
